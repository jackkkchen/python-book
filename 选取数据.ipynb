在数据分析中，选取需要的数据进行处理和分析是很重要的。Pandas 提供了丰富的数据探索方法，包数据排序、数据筛选、数据切片、数据分组汇总等等。这些功能可以帮助我们快速了解数据的基本特征并选取自己需要的数据。

在Pandas中，我们可以通过列名、位置索引以及各种条件来筛选感兴趣的数据子集，从而对数据进行过滤、筛选、切片、分析等操作。

首先，导入并查看泰坦尼克号数据集的前5行：

```python
# 从 github上的公开存储库获取数据
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv') 
# 查看前5行数据，加粗部分是列标签，第一列数字是行标签。
df.head()
```

## 筛选行数据

可以通过行索引获取行数据，冒号指定范围，一般的模式是，起始索引:结束索引，结束索引不含在内，开始索引为0或结束索引为行数时可省略数字。和NumPy的切片用法一样。

```python
# 选取前3行数据，行索引取0、1、2，结束索引3是取不到的。 
df[:3]  # 等价于df[0:3]  
# 选取第5行(索引4)到第8行(索引7) 
df[4:8] 
```

::: {.callout-warning}

注意在dataframe中索引0所表示的行才是第一行，所以切片的索引比行数小一个数。

:::


### iloc方法基于行索引筛选行

iloc方法是按行索引位置选取数据，可以引用索引列表(`[index1, index2, ...]`)或者索引区间(`start_index:stop_index`)。

```python
df.iloc[:4]
# 等价于df.iloc[:4, :]，选取行数据时可以不选取列，一个冒号表示取全部
```

### loc方法基于行标签筛选行

loc方法是按行标签选取数据，因为这里没有设置行标签，所以这里行标签默认等于位置索引。

```python
df.loc[[1, 4]]
# 等价于df.loc[[1, 4], :]
```

#### 除了列表形式，还可以用区间形式如a:b，引用连续的行标签。

```python
df.loc[:4]
# 等价于df.loc[0:4, :]
```

::: {.callout-warning}

注意使用`loc`方法是可以取到结束行的。

:::


```python
df_summary = df.describe()
df_summary.loc["count":"mean"]
```

## 选取列数据

通过列索引或列名可以获取DataFrame的单列数据或多列数据。

假如我想知道每位乘客的姓名（选取单列），`df`为数据表格，`Name`为列名。


### 引用列名提取单列

```python
df['Name']
df.Name
```

### loc和iloc方法提取单列

```python
# 使用.loc或.iloc取列: 逗号前筛选行,逗号后面筛选列
# 行数取全部时只用一个冒号就可以，列取Name列：
df.loc[:, 'Name']
# # 使用.iloc根据列索引来提取第4列(索引3)
df.iloc[:, 3]
```

### 提取多列数据

如果需要获取多列，比如所有乘客的幸存情况（`Name`、`Survived`），我们再添加一对中括号把所有的列括起来以获取多个列的数据，数据子集依旧为数据框dataframe。

下面的代码示例用.head()显示前5行。

```python
# 方法一：
df[['Name','Survived']]

# 方法二：
# df.loc[:, ['column_name1', 'column_name2', ...]]
df.loc[:, ['Name','Survived']]

# df.iloc[:, [column_index1, column_index2]]
df.iloc[:, [3, 1]].head()
```

::: {.callout-tip}

内部方括号定义了一个[带有列名或列索引的Python列表](https://docs.python.org/3/tutorial/datastructures.html#tut-morelists)，而外部方括号用于从 dataframe 中筛选数据。

**筛选行的时候可以不提取列，但取列的时候必须用":, "来选取行，单个冒号表示取全部行。**

:::


#### 除了列表形式，还可以用列标签区间提取连续的列，比如提取乘客的基本信息(Name、Sex、Age)：

```python
# 用区间提取连续的多个列
df.loc[:, "Name":"Age"].head() 
```

### 根据条件筛选行数据

提取十八岁以下的乘客信息:

```python
df[df["Age"] < 18].head()
```

### 根据多个条件筛选行数据
多个条件之间用`&`或`|`相连：`df[(条件1) & (条件2)]`
提取成年(age>=18)男性乘客信息:

```python
condition = (df.Age >= 18) & (df.Sex == "male")
df[condition] # 等价于 df[(df["Age"] >= 18) & (df["Sex"] == "male")]
```

在条件前加上`~`表示取反，即提取除了成年男性以外的乘客信息：

```python
df[~condition]
```

::: {.callout-warning}

多个条件之间进行连接时，只能用`&`表示和，用`|`表示或，不能用`and`或`or`。

:::

在条件中可以结合`isna()`和`notna()`来提取或删除缺失值所在的行：

```python
df = df[df['Age'].notna()]
df
```

DataFrames 可以通过多种方式进行过滤，其中最直观的是使用[布尔索引](https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#indexing-boolean)。

::: {.callout-tip}

这些对原始数据进行筛选和提取的操作并不会改变原数据df，如果需要保存数据子集，需要赋值给新的数据变量。

:::
